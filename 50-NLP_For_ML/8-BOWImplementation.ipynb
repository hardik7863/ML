```python
import pandas as pd
messages=pd.read_csv('SMSSpamCollection',sep='\t',names=["labels","message"])
```

```python
messages
```

```python
## Data Cleaning And Preprocessing
import re
# re -> regular expression
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review= re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## Create the Bag Of Words Model
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500)
# here max feature is bascially the first 2500 words which have highest frequency

# for binary BOW 
# cv=CountVectorizer(max_features=2500,binary=True)
```

```python
X=cv.fit_transform(corpus).toarray()
```

## **N-Gram**

```python
## Create The Bag Of Words Model with N-gram
from sklearn.feature_extraction.text import CountVectorizer
## for Binary Bow Enable binary=True
cv=CountVectorizer(max_features=200,binary=True,ngram_range=(2,3))
X=cv.fit_transform(corpus).toarray()
```
