## Data Cleaning

```python
## missing value
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted inti two sets based on Region from 122 th index, we can make a new column based on the region

1:"Bejaia region Dataset"

2:"Sidi-Bel Abbes Region Dataset"

Add new Column with region

```python
df=dataset
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df.info()
```

```python
df[["Region"]]=df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
### removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## remove the 122nd row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## Fix space in columns names
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

```python
## Changes the required columns as integer type
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
## Changing the other Columns to float datatype
objects=[features for features in df.columns if df[features].dtype=='O']
```

```python
for i in objects:
    if i !='Classes':
        df[i]=df[i].astype(float)

## since our output features is Classes hence we do not want to change its datatype 
# let it be object
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
# let have the cleaned  dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```
